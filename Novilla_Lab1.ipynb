Question 3: Determining Survival Rate
- The script calculates the survival rate using df["Survived"].mean() and prints the result.
- It allows for an analysis of what the survival rate indicates.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Total passengers
total_passengers = len(df)

# Number of survivors
num_survivors = df["Survived"].sum()

# Number of non-survivors
num_non_survivors = total_passengers - num_survivors

# Calculate survival rate
survival_rate = (num_survivors / total_passengers) * 100

# Print survival rate
print(f"Survival Rate: {survival_rate:.2f}%")

# Labels and values for visualization
labels = ["Survived", "Did Not Survive"]
values = [num_survivors, num_non_survivors]

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=["green", "red"])
plt.ylabel("Number of Passengers")
plt.title("Titanic Survival Count")
plt.show()

```

```python
import pandas as pd

# Load Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Filter survivors
survivors = df[df['Survived'] == 1]

# Extract ages and remove missing values
survivor_ages = survivors['Age'].dropna()

# Compute the average age
average_age_survivors = survivor_ages.mean()

print("Average Age of Survivors:", average_age_survivors)
```
